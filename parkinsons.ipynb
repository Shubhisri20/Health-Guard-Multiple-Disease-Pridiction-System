```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
parkinsons_data = pd.read_csv('parkinsons.csv')
parkinsons_data.head()
```

```python
parkinsons_data.shape
```

```python
parkinsons_data.describe()
```

```python
# describe  the target variable
parkinsons_data['status'].value_counts()
```

```python
print(parkinsons_data.dtypes)
```

```python
# Convert all columns that should be numeric
for column in parkinsons_data.columns:
    parkinsons_data[column] = pd.to_numeric(parkinsons_data[column], errors='coerce')
```

```python
numeric_columns = parkinsons_data.select_dtypes(include=['float64', 'int64']).columns
parkinsons_data_numeric = parkinsons_data[numeric_columns]
```

```python
result = parkinsons_data_numeric.groupby(parkinsons_data['status']).mean()
print(result)
```

```python
# grouping the data based on the target variable
parkinsons_data.groupby('status').mean()
```

```python
x = parkinsons_data.drop(columns=['name','status'] , axis=1)
y = parkinsons_data['status']
```

```python
print(x)
```

```python
print(y)
```

```python
#Splitting the data to training data and test data
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(x_train,x_test,y_train,y_test )
```

```python
model = svm.SVC(kernel = 'linear')
```

```python
# Training the SVM model with training data 
model.fit(x_train,y_train)
```

```python
# accuracy score on training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy score of the training data : ',training_data_accuracy)
```

```python
# accuracy score on the test data 
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy score of the test data : ',test_data_accuracy)
```
